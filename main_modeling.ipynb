# 모델링3. 통합점수 만들어 예측하기

- 순위 가점을 따로 계산하는 모델은 성능이 너무 떨어짐
- 이에 따라서 새롭게 통합점수를 계산해서 이를 예측하는 회귀 모델을 만들것이다
- 0.25가 일단 한계점인듯한데 0.25가 의미 하는것은

```python
trial_df=data_df.copy()

trial_df.head()
```

```python
# 통합점수 생성
# 순위와 가산점의 상대 중요도를 조정하는 가중치(alpha, beta)
alpha = 100.0  # 순위 점수에 대한 가중치
beta = 1.0   # 가산점에 대한 가중치

# 통합 점수 계산
# (4 - rate)는 "1순위 -> 3, 2순위 -> 2, 3순위 -> 1" 형태로 점수를 매기는 예시
# (score / 10)은 가산점(0~10점)을 0~1 범위로 정규화한 값
trial_df['Merged_score'] = alpha * (4 - trial_df['Cutline_rate']) + beta * (trial_df['Cutline_score'] / 10)

cols=['Latitude','Longitude','people','Infra_score']

trial_df=trial_df.drop(columns=cols)

trial_df.head()
```

- 새로운 로직 생각 해본거

통합점수를 만들때, 순위는 10자리수로 만들어서 예측한다

그리고 숫자가 0 ~5 라면 내리고, 6 ~10 이면 올려서 예측값을 조절해준다
- 이로직을 생각한 이유는 순위가 중요한 부분인데 한자리수 점수에서는 소수점 예측이 잘못되거나 하면 순위가 바뀌어버림
- 이 로직을 적용한 결과 : 순위에 대한것은 3개 정도만 제외하면 잘 들어 맞았다.
 > 그치만 데이터가 부족하니 단정짓지는 말자

- score에 대한 예측 정확도를 어떻게 올릴수있을까

```python
# 모델 예측값 반올림 함수
def custom_round_100(values):
    """
    1) 정수부(int_part)를 얻는다. (예: 225.7 -> 225)
    2) 소수부(dec_part)를 얻는다. (225.7 -> 0.7)
    3) '100의 자리'를 구한다. (225 -> 200)
    4) '나머지' remainder = int_part % 100 (예: 225 -> 25, 256 -> 56)
    5) remainder <= 50이면 아래쪽 100단위(200), remainder > 50이면 윗 100단위(300)로 조정
    6) 소수부는 그대로 더한다.
    """
    results = []
    for val in values:
        int_part = int(val)               # 정수부
        dec_part = val - int_part         # 소수부

        # 100단위로 내려놓기 (예: 225 -> 200, 256 -> 200)
        hundred_floor = (int_part // 100) * 100
        remainder = int_part % 100        # 예: 225 -> 25, 256 -> 56

        if remainder <= 50:
            # 0~50 사이는 아래 100단위
            new_int_part = hundred_floor
        else:
            # 51~99 사이는 위 100단위로 올림
            new_int_part = hundred_floor + 100

        new_val = new_int_part + dec_part
        results.append(new_val)

    return np.array(results)
```

```python
# 통합점수 역산 함수
def invert_merged_score(merged_val, alpha, beta):
    """
    Merged_score로부터 (rate, score) 후보를 찾는 함수.
    - rate ∈ {1, 2, 3}
    - score ∈ [0, 10]
    - 여러 해가 있을 수 있으므로, 우선 첫 번째 valid 해를 리턴.
    - 만약 어떤 rate로도 0~10 내 score가 안 나오면 (np.nan, np.nan) 반환.
    """
    # print(merged_val)
    # print('********************')
    for rate_candidate in [1, 2, 3]:
        # 수식에 따라 score를 역산
        score_candidate = (10 / beta) * (merged_val - alpha * (4 - rate_candidate))
        score_candidate=round(score_candidate)
        #print(score_candidate)
        # 유효 범위 체크
        if 0 <= score_candidate <= 10:
            return rate_candidate, score_candidate
    # print("***** 계산끝 *********")
    # 만약 어떤 rate도 0~10 범위 만족하는 해가 없다면
    return np.nan, np.nan
```

```python
# 데이터 분리
target='Merged_score'
x=trial_df.drop(columns=target)
x=x.drop(columns=['Cutline_rate','Cutline_score'])
y=trial_df[target]

# 라벨인코딩
# x['Supply_type'] = LabelEncoder().fit_transform(x['Supply_type'])
# 스케일링
scaler = MinMaxScaler()
x = scaler.fit_transform(x)
```

```python
# 데이터셋 분할
x_train,x_val,y_train,y_val=train_test_split(x,y,test_size=0.2,random_state=25)
```

## 기본 lgbm 모델

```python
# 모델링
model=LGBMRegressor(n_estimators=10)

model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_val)
y_pred=y_pred.round(1)
y_pred_custom=custom_round_100(y_pred)

print('MAE : ',mean_absolute_error(y_val,y_pred_custom))
print('MSE : ',mean_squared_error(y_val,y_pred_custom))
print('r2 score : ',r2_score(y_val,y_pred_custom))
```

```python
# 역산 적용하기
# 예측값을 역산하여 rate와 score 복원
predicted_values = pd.Series(y_pred_custom, index=y_val.index).apply(lambda m: invert_merged_score(m, alpha, beta))
predicted_rate_score = pd.DataFrame(predicted_values.tolist(), index=y_val.index, columns=['Predicted_rate', 'Predicted_score'])

# 원래 값과 예측값 병합
comparison_df = pd.concat([
    trial_df.loc[y_val.index, ['Cutline_rate', 'Cutline_score']],  # 원래 값
    predicted_rate_score  # 예측 값
], axis=1)

```

```python
# 평가 지표 계산 (rate는 분류, score는 회귀)
rate_accuracy = (comparison_df['Cutline_rate'] == comparison_df['Predicted_rate']).mean()
score_mae = mean_absolute_error(comparison_df['Cutline_score'], comparison_df['Predicted_score'])

print(f"Rate Accuracy: {rate_accuracy * 100:.2f}%")
print(f"Score MAE: {score_mae:.4f}")


print(confusion_matrix(comparison_df['Cutline_rate'], comparison_df['Predicted_rate']))

# Classification Report 출력
print("Classification Report for Rate Prediction:")
print(classification_report(comparison_df['Cutline_rate'], comparison_df['Predicted_rate'], digits=2))
```

## lgbm grid search

```python

model_lgb = LGBMRegressor(
    # 기본 파라미터
    random_state=42,
    n_estimators=100,  # 대략 크게 주고, early stopping으로 조절
    boosting_type='gbdt'
)

param_grid = {
    # 트리 구조 관련
    'num_leaves': [31, 63, 127],
    'max_depth': [-1, 5, 10],
    # 규제 항
    'reg_alpha': [0, 0.1, 1],
    'reg_lambda': [0, 0.1, 1],
    # 샘플링 관련
    'feature_fraction': [1.0, 0.8, 0.6],
    'bagging_fraction': [1.0, 0.8],
    # 학습률
    'learning_rate': [0.1, 0.01],
}

# -----------------------
# 2. GridSearchCV 설정
# -----------------------
grid_search = GridSearchCV(
    estimator=model_lgb,
    param_grid=param_grid,
    scoring='neg_mean_squared_error',  # 회귀에서는 MSE, RMSE 등 사용 가능
    cv=3,  # 3-Fold 교차 검증
    verbose=1,
    n_jobs=-1  # 멀티코어 사용
)

# -----------------------
# 3. fit() 시 early_stopping_rounds 적용
# -----------------------
grid_search.fit(x_train, y_train,)
```

## 모델링

```python
# [0] 예시: 데이터 준비
# trial_df 내의 'Cutline_rate', 'Cutline_score'를 타겟으로 분리
X = trial_df.drop(columns=['Cutline_rate', 'Cutline_score'])
y_rate = trial_df['Cutline_rate']
y_score = trial_df['Cutline_score']

# 범주형 피처(컬럼명 리스트)
cat_features = ['Supply_type']

# [1] Train-Test Split
X_train, X_test, y_rate_train, y_rate_test, y_score_train, y_score_test = train_test_split(
    X, y_rate, y_score,
    test_size=0.2,
    random_state=24
)

# [2] 분류 모델: CatBoostClassifier
classifier = CatBoostClassifier(
    iterations=1000,
    depth=8,
    learning_rate=0.1,
    loss_function='MultiClass',
    cat_features=cat_features,  # 범주형 컬럼
    verbose=0,
    one_hot_max_size=10,
    task_type='GPU'
)

# 분류 모델 학습
classifier.fit(X_train, y_rate_train)

# Train/Test 분류 예측
y_rate_pred_train = classifier.predict(X_train).astype(int)
y_rate_pred_test = classifier.predict(X_test).astype(int)

# [3] 분류 결과를 새 피처로 추가
X_train_reg = X_train.copy()
X_train_reg['pred_rate'] = y_rate_pred_train

X_test_reg = X_test.copy()
X_test_reg['pred_rate'] = y_rate_pred_test

# [4] 회귀 모델: CatBoostRegressor
regressor = CatBoostRegressor(
    iterations=1000,
    depth=8,
    learning_rate=0.1,
    loss_function='RMSE',
    cat_features=cat_features,
    verbose=0,
    task_type='GPU'
)

# 회귀 모델 학습
regressor.fit(X_train_reg, y_score_train)

# Test 데이터 예측
y_score_pred = regressor.predict(X_test_reg)

y_score_pred.round().astype(int)

# 결과 확인
print('Rate : ',classification_report(y_rate_test,y_rate_pred_test))
print('Rate CM:',confusion_matrix(y_rate_test,y_rate_pred_test))
print('Score MAE:',mean_absolute_error(y_score_test,y_score_pred))
print('Score R2:',r2_score(y_score_test,y_score_pred))
```
